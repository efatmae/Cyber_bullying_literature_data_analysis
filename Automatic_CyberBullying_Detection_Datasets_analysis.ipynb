# Data annotation process

### crowd sourcing

```python
data_annotation_process = ref_df[ref_df["Data Collection process"] != "no"]
```

```python
len(data_annotation_process)
```

```python
data_crowdsourced = data_annotation_process[(data_annotation_process["Data Collection process"].str.contains("Crowd") | 
                                   data_annotation_process["Data Collection process"].str.contains("AMT")|
                                   data_annotation_process["Data Collection process"].str.contains("crowd")|
                                    data_annotation_process["Data Collection process"].str.contains("CF"))|
                                    data_annotation_process["Data Collection process"].str.contains("Figure")]
```

```python
len(data_crowdsourced)
```

```python
data_crowdsourced.to_csv("Data/data_crowdsourced.csv")
```

```python
data_crowdsourced[["Dataset","Data Collection process"]]
```

### annotated locally

```python
data_not_crowdsourced = data_annotation_process[(data_annotation_process["Data Collection process"].str.contains("Crowd")== False) & 
                                   (data_annotation_process["Data Collection process"].str.contains("AMT")==False) &
                                   (data_annotation_process["Data Collection process"].str.contains("crowd")==False)&
                                   (data_annotation_process["Data Collection process"].str.contains("CF")==False)&
                                   (data_annotation_process["Data Collection process"].str.contains("Figure")==False)]
```

```python
data_not_crowdsourced_annotated = data_not_crowdsourced[(data_not_crowdsourced["Data Collection process"].str.contains("annotated")== True) | 
                                   (data_not_crowdsourced["Data Collection process"].str.contains("Annotated")==True) |
                                   (data_not_crowdsourced["Data Collection process"].str.contains("labelled")==True)|
                                   (data_not_crowdsourced["Data Collection process"].str.contains("anntated")==True)|
                                  (data_not_crowdsourced["Data Collection process"].str.contains("labeled")==True)|
                                   (data_not_crowdsourced["Data Collection process"].str.contains("Labelled")==True)]
```

```python
len(data_not_crowdsourced_annotated)
```

```python
data_not_crowdsourced_annotated.to_csv("Data/data_not_crowdsourced_annotated_locally.csv")
```

```python
data_not_crowdsourced_annotated[["Dataset","Data Collection process"]]
```

### not labelled by paper

```python
data_not_crowdsourced_not_annotated_by_researchers= data_annotation_process[
    ~data_annotation_process["Data Collection process"].isin(data_crowdsourced["Data Collection process"])&
~data_annotation_process["Data Collection process"].isin(data_not_crowdsourced_annotated["Data Collection process"])]
```

```python
len(data_not_crowdsourced_not_annotated_by_researchers)
```

```python
data_not_crowdsourced_not_annotated_by_researchers.to_csv("Data/data_not_crowdsourced_not_annotated_by_researchers.csv")
```
